# CALCULATE EACH CROPS'S SHARE OF TOTAL DISTRICT AREA



```python
crop_area = df1.groupby(['District_Name','Crop'])['Area'].sum().reset_index()

district_total = df1.groupby('District_Name')['Area'].sum().reset_index()
district_total.columns = ['District_Name','Total_District_Area']

crop_area = crop_area.merge(district_total, on='District_Name')
crop_area['Share_%'] = (crop_area['Area'] / crop_area['Total_District_Area']) * 100

crop_area.sort_values(['District_Name','Share_%'], ascending=[True, False])
```

# IDENTIFY THE DOMINANT CROP PER DISTRICT

```python
dominant_crop = crop_area.loc[crop_area.groupby('District_Name')['Area'].idxmax()]
dominant_crop = dominant_crop[['District_Name','Crop','Area','Share_%']]
dominant_crop
```

```python
crop_prod = df1.groupby(['District_Name','Crop'])['Production'].sum().reset_index()
dominant_crop_prod = crop_prod.loc[crop_prod.groupby('District_Name')['Production'].idxmax()]
dominant_crop_prod
```

# CATEGORZE DISTRICTS INTO RISK BANDS (DIVERSIFIED / MODERATE / CONCENTRATED)



```python
def risk_band(share):
    if share < 30:
        return 'Diversified'
    elif share < 50:
        return 'Moderate'
    else:
        return 'Concentrated'

dominant_crop['Risk_Band'] = dominant_crop['Share_%'].apply(risk_band)
dominant_crop
```

#  YEAR-OVER-YEAR TREND — GROUP BY DISTRICT , CROP YEAR AND CHECK IF PRODUCTION/YIELD IS RISING OR FALLING


```python
year_trend = df1.groupby(['District_Name','Crop_Year'])[['Production','Yield']].sum().reset_index()
year_trend = year_trend.sort_values(['District_Name','Crop_Year'])

year_trend['Production_Change_%'] = year_trend.groupby('District_Name')['Production'].pct_change() * 100
year_trend['Yield_Change_%'] = year_trend.groupby('District_Name')['Yield'].pct_change() * 100

year_trend
```

```python
district_year = df1.groupby(['District_Name','Crop_Year'])['Production'].sum().reset_index()


# District-level average & std
district_stats = district_year.groupby('District_Name')['Production'].agg(['mean','std']).reset_index()
district_stats.columns = ['District_Name','Avg_Production','Std_Production']

district_year = district_year.merge(district_stats, on='District_Name')

# Z-score for each district-year
district_year['Z_Score'] = (district_year['Production'] - district_year['Avg_Production']) / district_year['Std_Production']

# % deviation from district average
district_year['Pct_Deviation'] = ((district_year['Production'] - district_year['Avg_Production']) / district_year['Avg_Production']) * 100

# Flag anomalies: sharp drop = z-score below -1.5 (or -2 for stricter)
district_year['Anomaly'] = district_year['Z_Score'] < -1.5

anomalies = district_year[district_year['Anomaly']]
anomalies.sort_values('Z_Score')
```
